```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate random points (x, y) between -1 and 1
x = 2 * np.random.rand(1000) - 1
y = 2 * np.random.rand(1000) - 1
```

## My scatter plot

```python
plt.style.use('ggplot')

plt.scatter(x, y, color='y')

# Add a circumference to count inner points
circumference = plt.Circle((0, 0), 1, color='b', fill=False)
plt.gca().add_patch(circumference)
```

```python
# Counts points inside the circumference
inner_points = 0
for point in zip(x, y):
    if point[0] ** 2 + point[1] ** 2 <= 1:
        inner_points += 1
```

```python
# Computes an approximated value of pi using the Monte Carlo formula
pi = 4 * inner_points / len(x)
print("An approximation of pi: ", pi)
```
